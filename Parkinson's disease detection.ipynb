```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**Data Analysis**

```python
df_parkinsons = pd.read_csv('parkinsons.csv')
```

**1 - Parkinson's Positive**

**0 - Negative**

```python
df_parkinsons = df_parkinsons.drop(columns='name',axis=1)
```

```python
df_parkinsons.groupby('status').mean()
```

**Data Preprocessing**

```python
X = df_parkinsons.drop(columns = ['name','status'],axis = 1)
Y = df_parkinsons['status']
```

```python
#splitting the data into train and test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state =2 )
```

**Data Standardization**

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

**MODEL BUILDING**

```python
model = svm.SVC(kernel = 'linear')
```

```python
model.fit(X_train, Y_train)
```

```python
#MODEL EVALUATION
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_pred)
```

```python
print('Accuracy score of train data :', training_data_accuracy)
```

```python
X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test,X_test_pred)
```

```python
print('Accuracy score of test data :', testing_data_accuracy)
```

**Building a predictive system**

```python
import numpy as np

# Input data as a list
input_data = [
    203.18400,  # MDVP:Fo(Hz)
    211.52600,  # MDVP:Fhi(Hz)
    196.16000,  # MDVP:Flo(Hz)
    0.00178,    # MDVP:Jitter(%)
    0.000009,   # MDVP:Jitter(Abs)
    0.00094,    # MDVP:RAP
    0.00106,    # MDVP:PPQ
    0.00283,    # Jitter:DDP
    0.00958,    # MDVP:Shimmer
    0.08500,    # MDVP:Shimmer(dB)
    0.00468,    # Shimmer:APQ3
    0.00610,    # Shimmer:APQ5
    0.00726,    # APQ
    0.01403,    # DDA
    0.00065,    # NHR (Noise-to-Harmonics Ratio)
    33.04700,   # HNR (Harmonics-to-Noise Ratio)
    0.340068,   # RPDE
    0.741899,   # DFA
    -7.964984,  # spread1
    0.163519,   # spread2
    1.423287,   # D2
    0.044539    # PPE
]

# Changing the input data to numpy array
input_data_numpy_array = np.asarray(input_data)

# Reshape the numpy array
input_data_reshape = input_data_numpy_array.reshape(1, -1)

# Standardize the data
std_data = scaler.transform(input_data_reshape)

# Make prediction
prediction = model.predict(std_data)
print(prediction)

if prediction[0] == 0:
    print('The person does not have Parkinson\'s disease')
else:
    print('The person has Parkinson\'s disease')
```
